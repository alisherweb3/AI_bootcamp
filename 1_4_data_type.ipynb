## Tipizatsiyalash

```python
variable1 = input("Birinchi raqam ")
variable2 = input("Ikkinchi raqam ")

print(variable1 + variable2)
```

Bu holatda bizga maxsus funksiyalar yordam beradi, har bir type uchun alohida
* str() — "" qo'shtirnoq ichidagi argumentni, string(matn) tarzda qaytaradi;
* int() - "" qo'shtirnoq ichidagi argumentni, integer(raqam) tarzda qaytaradi;
* float() — "" qo'shtirnoq ichidagi argumentni, .00 (10 butun) tarzda qaytaradi.

agar siz funksiyaga kiritilgan data, qanday ekanligiga amin bo'lmasangiz:
* type() — "" qo'shtirnoq ichigadi argumentni data type ni chiqarib beradi;

```python
print(type(variable1))
```

```python
variable1 = input("birinchi o'zgaruvchini kiriting ")
variable1 = int(variable1)

variable2 = input("ikkinchi o'zgaruvchini kiriting ")
variable2 = int(variable2)


print(variable1 + variable2)
```

Dasturni yana **qisqartirish** mumkin.
